```python
import pandas as pd
import requests
from io import StringIO
```

```python
#getting the table from the website
url = "https://en.wikipedia.org/api/rest_v1/page/html/List_of_most-visited_museums"

headers = {
    'User-Agent': 'Mozilla/5'
    }

response = requests.get(url, headers = headers).text
museums = pd.read_html(StringIO(response))
museums_df = museums[0]
museums_df
```

```python
museums_df["Visitors"]
```

```python
museums_df["Visitors_2023_2024"] = museums_df["Visitors"].str.split(r"[/(/[]", n=1).str[0]
museums_df.head(10)
```

```python
museums_df.drop(columns= ["Visitors"], inplace = True)
```

```python
museums_df
```

```python
museums_df["Address_1"] = museums_df["Name"] + "," + museums_df['City'] + "," + museums_df['Country']
museums_df
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
geolocator = Nominatim(user_agent= 'User_Marv', timeout= 10)
geocoder = RateLimiter(geolocator.geocode, min_delay_seconds= 1)
```

```python
#Creating new columns to save geocoded info
museums_df['address'] = 0
museums_df['lon'] = 0
museums_df['lat'] = 0
museums_df
```

```python
#function to geocode
def geocoding():
  for index, row in museums_df.iterrows():
    #print(f"{index + 1}. {row['Address_1']}")
    addr = row["Address_1"]
    location = geocoder(addr)

    try:
      address_2 = location.address
      lon = location.longitude
      lat = location.latitude
      #saving the address to the df
      museums_df.loc[index, "address"]= address_2
      museums_df.loc[index, "lon"]= lon
      museums_df.loc[index, "lat"]= lat

    except AttributeError:
      museums_df.loc[index, "address"]= 'No info'
      museums_df.loc[index, "lon"]= 'No info'
      museums_df.loc[index, "lat"]= 'No info'

    print(f"Now geocoding {index + 1}. {addr}")
geocoding()
```
